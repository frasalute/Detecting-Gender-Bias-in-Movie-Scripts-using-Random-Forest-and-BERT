## Pre-processing

```python
# extend stopwords list
stop_words = set(stopwords.words('english'))
stop_words.update([',', '.', '?', ':', ';', "'", '“', '”', '!', '’', '...', '....', '--', '_'])

# initialize lemmatizer and stemmer
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\w\s]', '', text)
    words = word_tokenize(text)
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    words = [stemmer.stem(word) for word in words if word not in stop_words]
    return ' '.join(words)
```

```python
# apply preprocessing
final_df['processed_text'] = final_df['text'].apply(preprocess_text)
```

```python
vectorizer = TfidfVectorizer(max_features=5000) 
X_text = vectorizer.fit_transform(final_df['processed_text'])
```

```python
le = LabelEncoder()
final_df['gender_label'] = le.fit_transform(final_df['meta.gender'])
y = final_df['gender_label']
```

Train the Model

```python
# combine the text features with the other features
X_other = final_df[['line_count', 'encoded_credit_pos', 'medianLineLength', 'medianWordCountLine']]

# convert X_other to a sparse matrix and combine with X_text
X_other_sparse = csr_matrix(X_other.values)
X_combined_1 = hstack([X_text, X_other_sparse])
```

```python
# split the data into 80% training and 20% testing sets
X_train, X_test, y_train, y_test = train_test_split(X_combined_1, y, test_size=0.2, random_state=42)
```

## Random Forest with BERT

```python
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')

def get_bert_embeddings(text):
    inputs = bert_tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        outputs = bert_model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

# Apply BERT embeddings to the processed text column
final_df['bert_embeddings'] = final_df['processed_text'].apply(get_bert_embeddings)
bert_embeddings_array = np.vstack(final_df['bert_embeddings'].values)
```

```python
# Combine with other features because just BERT might not capture everything important
X_other = final_df[['line_count', 'encoded_credit_pos', 'medianLineLength', 'medianWordCountLine']].values
X_combined = np.hstack([X_other, bert_embeddings_array])
```

```python
start_time = time.time()

X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)

rf_bert = RandomForestClassifier(
    n_estimators=500,  
    max_depth=20, 
    min_samples_split=5, 
    min_samples_leaf=2, 
    random_state=42,
    n_jobs=-1 
)
rf_bert.fit(X_train, y_train)


y_pred = rf_bert.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print(classification_report(y_test, y_pred))

end_time = time.time()
forest_time_bert = end_time - start_time
print(f"Total Execution Time: {forest_time_bert} seconds")
```

```python
# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

## Logistic Regression with BERT

```python
start_time = time.time()

X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)

# Using adjustment of parameter C to control regularization 
for C_value in [0.01, 0.1, 1, 10, 100]:
    lr_bert = LogisticRegression(C=C_value, max_iter=10000, random_state=42)
    lr_bert.fit(X_train, y_train)
    y_pred = lr_bert.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    print(f"C = {C_value} | Accuracy: {accuracy:.2f}")

print(classification_report(y_test, y_pred))


end_time = time.time()
log_reg_bert = end_time - start_time
print(f"Total Execution Time: {log_reg_bert} seconds")
```
